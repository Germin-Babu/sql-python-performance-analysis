# Create an aggregate table for our analysis

```python
freight_summary = pd.read_sql_query("""SELECT VendorNumber, SUM(Freight) as FreightCost FROM vendor_invoice
                                    GROUP BY VendorNumber""", conn)
freight_summary                                    
```

```python
pd.read_sql_query("""SELECT
    p.VendorNumber,
    p.VendorName,
    p.Brand,
    p.PurchasePrice,
    price.Volume,
    price.Price as ActualPrice,
    SUM(p.Quantity) as TotalPurchaseQuantity,
    SUM(p.Dollars) as TotalPurchaseDollars
    FROM purchases p
    JOIN purchase_prices price
    ON p.Brand = price.Brand
    WHERE p.PurchasePrice > 0
    GROUP BY p.VendorNumber, p.VendorName, p.Brand
    ORDER BY TotalPurchaseDollars""", conn)
```

```python
pd.read_sql_query("""SELECT
    VendorNo,
    Brand,
    SUM(SalesDollars) as TotalSalesDollars,
    SUM(SalesPrice) as TotalSalesPrice,
    SUM(SalesQuantity) as TotalSalesQuantity,
    SUM(ExciseTax) as TotalExciseTax
    FROM sales
    GROUP BY VendorNo, Brand
    ORDER BY TotalSalesDollars""", conn)
```

```python
vendor_summary = pd.read_sql_query("""WITH FreightSummary AS (
    SELECT
        VendorNumber,
        SUM(Freight) AS FreightCost
    From vendor_invoice
    GROUP BY VendorNumber
),
PurchaseSummary AS (
    SELECT 
    p.VendorNumber,
    p.VendorName,
    p.Brand,
    p.Description,
    p.PurchasePrice,
    price.Price as ActualPrice,
    price.Volume,
    SUM(p.Quantity) as TotalPurchaseQuantity,
    SUM(p.Dollars) as TotalPurchaseDollars
    FROM purchases p
    JOIN purchase_prices price
    ON p.Brand = price.Brand
    WHERE p.PurchasePrice > 0
    GROUP BY p.VendorNumber, p.VendorName, p.Brand, p.Description, p.PurchasePrice, price.Price, price.Volume
),
SalesSummary AS (
    SELECT
    VendorNo,
    Brand,
    SUM(SalesDollars) as TotalSalesDollars,
    SUM(SalesPrice) as TotalSalesPrice,
    SUM(SalesQuantity) as TotalSalesQuantity,
    SUM(ExciseTax) as TotalExciseTax
    FROM sales
    GROUP BY VendorNo, Brand
)
SELECT
    ps.VendorNumber,
    ps.VendorName,
    ps.Brand,
    ps.Description,
    ps.PurchasePrice,
    ps.ActualPrice,
    ps.Volume,
    ps.TotalPurchaseQuantity,
    ps.TotalPurchaseDollars,
    ss.TotalSalesDollars,
    ss.TotalSalesPrice,
    ss.TotalSalesQuantity,
    ss.TotalExciseTax,
    fs.FreightCost
FROM PurchaseSummary ps
LEFT JOIN SalesSummary ss
    ON ps.VendorNumber = ss.VendorNo
    AND ps.Brand = ss.Brand
LEFT JOIN FreightSummary fs
    ON ps.VendorNumber = fs.VendorNumber
ORDER BY ps.TotalPurchaseDollars DESC""", conn)
```

```python
vendor_summary['Volume'] = vendor_summary['Volume'].astype('float64')
```

```python
vendor_summary.fillna(0, inplace =True)
```

```python
#Removes whitespace
vendor_summary['VendorName'].str.strip()
```

```python
vendor_summary['GrosProfit'] = vendor_summary['TotalSalesDollars'] - vendor_summary['TotalPurchaseDollars']
```

```python
vendor_summary['ProfitMargin'] = (vendor_summary['GrosProfit'] / vendor_summary['TotalSalesDollars']) * 100
```

```python
vendor_summary['StockTurnover'] = vendor_summary['TotalSalesQuantity']/vendor_summary['TotalPurchaseQuantity']
```

```python
vendor_summary['SalestoPurchaseRatio'] = vendor_summary['TotalSalesDollars'] / vendor_summary['TotalPurchaseDollars']
```

```python
cursor = conn.cursor()
cursor.execute("""CREATE TABLE vendor_summary(
     VendorNumber INT,
     VendorName VARCHAR(100),
     Brand INT,
     Description VARCHAR(100),
     PurchasePrice DECIMAL (15,2),
     ActualPrice DECIMAL (15,2),
     Volume,
     TotalPurchaseQuantity INT,
     TotalPurchaseDollars DECIMAL (15,2),
     TotalSalesDollars DECIMAL (15,2),
     TotalSalesPrice DECIMAL (15,2),
     TotalSalesQuantity INT,
     TotalExciseTax DECIMAL (15,2),
     FreightCost DECIMAL (15,2),
     GrosProfit DECIMAL (15,2),
     ProfitMargin DECIMAL (15,2),
     StockTurnover DECIMAL (15,2),
     SalestoPurchaseRatio DECIMAL (15,2),
     PRIMARY KEY (VendorNumber, Brand)
);
""")
```
